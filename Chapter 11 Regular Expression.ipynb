 Regular expression handles taks more elegantly. Own little programming language for searching and parsing strings.

```python
import re
```

```python
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    if re.search('From:',line):#to search a line with"From:"
        print(line)
```

```python
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    if re.search('^From:',line):#"^"mmeans to find the line start with "From:"
        print(line)
```

```python
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    if re.search('^From:',line):#to find the line start with "Fxxm:"
        print(line)
```

```python
#wild card character to narrow down the match lines
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    if re.search('^From:.+@',line):#to find the line start with "From:" followed by one or more characters".+" followed by an @ sign
        print(line)
```

```python
#extracting dtat using regular expressions
s='Hello from csev@umich.edu to cwen@iupui.edu about the meeting @2PM'
lst=re.findall('\S+@+\S+',s)#findall extract one or more.'+s'means non-whitespace character
print(lst)
```

```python
#It may cause "greedy" matching which means match more than you want
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    x=re.findall('\S+@\S+',line)
    if len(x)>0:
        print(x)
```

```python
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    x=re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z]',line)#narrow down the request the non-whitespace before are characters or numbers only
    if len(x)>0:
        print(x)
```

```python
#Combining searching and extracting
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    if re.search('^X\S*: [0-9.]+',line):#to find a line starts with X followed by non-whitespace and ":",a withe space and numbers
        print(line)
```

```python
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    x=re.findall('^X\S*: ([0-9.]+)',line)#using findall instead of search using"()"to extract the value wanted
    if len(x)>0:
        print(x)
```

```python
#extract all of revision numbers
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    x=re.findall('^Details:.*rev=([0-9.]+)',line)#looking for lines start with"Detailes:" followed by any number of characters".*"
    if len(x)>0:
        print(x)
```

```python
#to extract hour eg. From xxx@xxxx Sat Jan 5 09:14:16 2008 take 09 out
hand=open('mbox-short.txt')
for line in hand:
    line=line.rstrip()
    x=re.findall('^From .* ([0-9.][0-9]):',line)#be careful with space it is brittle code that assuming thelines are nicely formatted
    if len(x)>0:
        print(x)
```
